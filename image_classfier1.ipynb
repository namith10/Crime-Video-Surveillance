```python
classes=['covid','normal','virus']
```

```python
def create_dataset():
    image_data=[]
    classes=['covid','normal','virus']
    for folder_index,folder_name in enumerate(classes):
        image_path=f'E:/datasets/Covid image dataset/archive (1)/COVID_IEEE/{folder_name}'
        image_path_os=os.listdir(image_path)
        for file_index,file_name in enumerate(image_path_os):
            path=os.path.join(image_path,file_name)
            image=Image.open(path)
            image1=image.resize((20,20))
            data=asarray(image1)
            image_data.append(data)
    # image_data=np.asarray(image_data, dtype='object')
    return(image_data)
image_data=create_dataset()
```
